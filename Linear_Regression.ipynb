# Basic Understanding of Linear Regression:

**Linear Regression** is a statistical method for making a model between Dependent or target value and Independent features.

slope-intercept-form.png

```python
#import Library functions

import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#Prepare some data

x = np.array([5, 20, 30, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 15, 20, 32, 22, 38])
```

**Graph Presentation**

```python
plt.scatter(x, y, label= "Points", color= "green", marker= ".", s= 100)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Graph for Linear Regression")
```

```python
# Print x value
x
```

```python
print(x.shape)
```

```python
print(y)
print(y.shape)
```

Create Model and Fit it

```python
model = LinearRegression()
```

```python
model.fit(x, y)
```

calculate the value of the weights b0 and b1,

**Get the Result**

```python
r_sq = model.score(x, y)
print(f"Coefficient : {r_sq}")
```

```python
print(f"Intercept: {model.intercept_}")

print(f"Slope: {model.coef_}")
```

```python
plt.scatter(x, y, label= "Points", color= "green", marker= ".", s= 100)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Graph for Linear Regression")
```

**Predict Response**

```python
y_pred = model.predict(x)
print(f"Predicted Response: \n{y_pred}")
```

**Mannual Prediction**

```python
y_pred = model.intercept_ + model.coef_*x
print(f"Predicted Response: \n{y_pred}")
```

# **Multiple Linear Regression**

```python
x_mult = [[0, 1], [5, 1], [15, 2], [25, 5], [35, 11], [45, 15], [55, 34], [60, 35]]
y_mult = [4, 5, 20, 14, 32, 22, 38, 43]

x_mult, y_mult = np.array(x_mult), np.array(y_mult)
```

```python
print(x_mult)
print(y_mult)
```

**Model Preparation**

```python
model = LinearRegression().fit(x_mult, y_mult)
```

**Get Result**

```python
print(f"Intercept: \n{model.intercept_}")
print(f"Coefficient: \n{model.coef_}")
```

**Prediction**

```python
y_pred = model.predict(x_mult)
print(f"Predicted Response: \n{y_pred}")
```

**Mannual Prediction**

```python
y_pred = model.intercept_ + np.sum(model.coef_ * x_mult, axis = 1)
print(f"Predicted Response: \n{y_pred}")
```

```python
plt.scatter(y_pred, y_mult, label= "Points", color= "green", marker= ".", s= 30)
plt.xlabel("Y Prediction")
plt.ylabel("Y")
plt.title("Graph for Prediction of Multiple Linear Regression")
```
